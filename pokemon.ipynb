```python
import pandas as pd
```

```python
csv = pd.read_csv('./pokemon.csv')
csv
```

```python
no_second_type = csv['Type 2'].isna()
pokemon_with_single_type = csv[no_second_type][['#', 'Name', 'Type 1']]
pokemon_with_single_type
```

```python
second_type_filter= csv['Type 2'].notna()
pokemon_with_second_type = csv[second_type_filter][['#', 'Name', 'Type 1', 'Type 2']]
pokemon_with_second_type
```

```python
legendary_filter = (csv['Legendary'] == True) & (csv['Generation'] == 5)
legendary_pokemon = csv[legendary_filter][['#', 'Name', 'Type 1', 'Type 2']]
legendary_pokemon['Type 2'] = legendary_pokemon['Type 2'].fillna('-')
legendary_pokemon
```
